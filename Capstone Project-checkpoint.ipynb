```python
encoder = OneHotEncoder(sparse_output = False, handle_unknown = 'ignore')

X_train_encoded = encoder.fit_transform(X_train[categorical_col])

X_test_encoded = encoder.fit_transform(X_test[categorical_col])
```

### Conclusion

The Logistic Regression model (Balanced) demonstrated the best performance in terms of both accuracy and recall, achieving an accuracy of 89%. However, it also produced 2,031 false positives for diabetes and only 200 false negatives.

Given the critical nature of minimizing false negatives for diabetes detection, I am prioritizing models with the highest recall and the lowest false negative rates to ensure accurate identification of patients at risk.

### Recommendation

I recommend that the company carefully monitor false negative predictions for diabetes, as the Logistic Regression model (Balanced) reported 200 cases where diabetes was not accurately identified. To address the potential risks associated with these false negatives, I propose implementing a physical verification process for all negative predictions. This step is vital because the impact of a patient incorrectly believing they do not have diabetes can lead to severe health consequences. In contrast, a false positive result can be resolved through additional confirmatory testing, minimizing its overall impact.
